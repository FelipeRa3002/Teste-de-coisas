### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
import matplotlib.pyplot as plt
P_loja1 = sum(loja['Preço'])
P_loja2 = sum(loja2['Preço'])
P_loja3 = sum(loja3['Preço'])
P_loja4 = sum(loja4['Preço'])
Fatruramneto = [P_loja1, P_loja2, P_loja3, P_loja4]
plt.bar(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'], Fatruramneto)
plt.title('Faturamento das Lojas em milhões')
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.show()
loja_f = "Loja 1" if P_loja1 > P_loja2 and P_loja1 > P_loja3 and P_loja1 > P_loja4 else "Loja 2" if P_loja2 > P_loja1 and P_loja2 > P_loja3 and P_loja2 > P_loja4 else "Loja 3" if P_loja3 > P_loja1 and P_loja3 > P_loja2 and P_loja3 > P_loja4 else "Loja 4"
loja_n = "Loja 1" if P_loja1 < P_loja2 and P_loja1 < P_loja3 and P_loja1 < P_loja4 else "Loja 2" if P_loja2 < P_loja1 and P_loja2 < P_loja3 and P_loja2 < P_loja4 else "Loja 3" if P_loja3 < P_loja1 and P_loja3 < P_loja2 and P_loja3 < P_loja4 else "Loja 4"
print(f"A loja com maior faturamento é {loja_f} e com menor {loja_n}")
```

# 2. Vendas por Categoria

# 3. Média de Avaliação das Lojas

```python
import matplotlib.pyplot as plt
Avalicacoes = [loja['Avaliação da compra'], loja2['Avaliação da compra'], loja3['Avaliação da compra'], loja4['Avaliação da compra']]
plt.hist(Avalicacoes, edgecolor='black')
plt.title('Avaliação da compra')
plt.xlabel('Avaliação')
plt.ylabel('Quantidade')
plt.legend()
plt.show()
#Loja 1 Azul, Loja 2 Laranja , Loja 3 Verde e Loja 4 Vermelho
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt
p = loja.groupby('Produto').size()
plt.barh(p.index, p.values)
plt.title('Produtos Mais Vendidos')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.show()
```

# 5. Frete Médio por Loja


```python
l = loja['Frete']
l = l.sum()/len(l)
l2 = loja2['Frete']
l2 = l2.sum()/len(l2)
l3 = loja3['Frete']
l3 = l3.sum()/len(l3)
l4 = loja4['Frete']
l4 = l4.sum()/len(l4)
Frete = [l, l2, l3, l4]
plt.bar(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'], Frete)
plt.title('Frete Médio das Lojas')
plt.xlabel('Lojas')
plt.ylabel('Frete')
plt.show()
```
